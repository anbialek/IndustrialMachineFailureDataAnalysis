## 5. Data Transformation

For the purposes of Exploratory Data Analysis (EDA), we add some new features derived from the previously created **Cycle ID**:

- **Cycle ID**: Based on the tool replacement events identified in the **Tool wear [min]** column (records equal to 0), we determine individual cycles. This allows us to estimate tool lifetime during the production process.

- **Temperature difference [K]**: We calculate the difference between process and air temperatures. Any anomalies (e.g., negative gradients) may indicate process instability or sensor measurement errors.

- **Process temperature step [K]**: Defined as the difference between consecutive process temperature readings within each product cycle. This feature reflects how quickly the process heats up or cools down, which may indicate abnormal operating conditions or potential sensor faults.

- **Air temperature step [K]**: Analogous feature calculated for the air sensor. Although temperature typically varies slowly, this step allows verification of sensor stability and detection of environmental anomalies.

- **Temperature step difference [K]**: Computed as the difference between process and air temperature steps. This represents the relative rate of thermal change between the process and its environment.

- **Power [kW]**: To better capture the mechanical load, we create a new variable combining Torque [Nm] and Rotational speed [rpm]:
$$
P = \frac{T \, n \, 2\pi}{60}
$$

and convert it to kW by dividing by 1000. This feature expresses the actual mechanical power delivered by the spindle, providing a more interpretable and potentially predictive indicator of operating conditions than torque or speed alone.

```python
df.head(3)
```

```python
df['Temperature difference [K]'] = df['Process temperature [K]']- df['Air temperature [K]']
```

```python
df['Air temperature step [K]'] = df.groupby('Cycle ID')['Air temperature [K]'].diff()
```

```python
df['Process temperature step [K]'] = df.groupby('Cycle ID')['Process temperature [K]'].diff()
```

```python
df['Temperature step difference [K]'] = df['Process temperature step [K]']- df['Air temperature step [K]']
```

```python
df['Power [kW]'] = round((2 * np.pi * df['Torque [Nm]'] * df['Rotational speed [rpm]'] )/ (60  * 1000), 2)
```

## 7. Key Findings / Insights

Our analysis shows that the environment and process temperature settings do not have a significant influence on tool wear, nor do they provoke machine breakdowns — no abnormal thermal variations were detected. The stable temperature profiles confirm that the cooling and thermal control systems function effectively and are not a source of instability.

The main process parameters — rotational speed (≈1500 rpm) and torque (≈40 Nm) — represent safe operating conditions, as machine failures are rarely observed within this range. However, lower spindle speeds are associated with higher machine load, as confirmed by the power and torque trends. Under these conditions, the process becomes less stable, which may lead to material fatigue failures and unplanned production downtime.

In terms of tool wear, a longer lifetime (above ≈100 min) appears to increase the risk of machine breakdown. To reduce the likelihood of production stops caused by excessive tool degradation, it is recommended to implement an additional operator inspection once tool wear exceeds this threshold or consider using a more wear-resistant tool material suitable for the operating conditions.

Additionally, failures occur more frequently for low-quality (L-type) products. This suggests that process parameters optimized for standard products may not be fully appropriate for low-quality batches, leading to increased mechanical stress and a higher probability of failure.

## 8. Conclusion & Next Steps

The conducted analysis provided valuable insights into the production process, helping to identify the main risk factors leading to machine failures and potential areas for optimization. The findings highlight the influence of mechanical load and process configuration on equipment reliability and product quality.

Based on these results, the following improvement actions are recommended:

- **Prevent excessive load conditions:**
exclude or limit low-speed spindle operation ranges, or introduce automated warning thresholds to prevent high torque levels and mechanical overload.

- **Optimize process settings for L-type products:**
review and adjust process parameters for low-quality (L-type) batches to improve process stability and reduce the frequency of machine breakdowns.

- **Enhance tool wear monitoring:**
introduce additional operator inspection or automatic alerts once tool wear exceeds approximately 100 minutes to prevent failure due to tool degradation.

- **Develop a predictive maintenance model:**
implement a machine learning model to forecast potential failures based on mechanical parameters (torque, speed, power, and tool wear), enabling proactive maintenance planning.

- **Strengthen quality assurance in risk zones:**
introduce additional product quality inspection for items produced under high-risk operating conditions (e.g., high torque, prolonged tool wear) to minimize the risk of delivering nonconforming products to customers.

- **Integrate findings into OEE improvement strategy:**
continue monitoring correlations between mechanical load, tool wear, and product type to refine process efficiency and overall equipment effectiveness (OEE).
